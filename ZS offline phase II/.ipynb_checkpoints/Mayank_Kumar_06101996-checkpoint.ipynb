```python
# Let's undergo a univariate analysis of all of 
# our features to better understand the distribution 
# of our features
# ======================================================================================================
# ======================================================================================================
#**************************************************************************************************************************	
	
	
#  different types of utility functions used for various univariate analysis (monks advanced method for univariate analysis)
from matplotlib import pyplot as plt
import seaborn as sns


def plot(x=None, y=None, hue=None, data=None,name='',barplot=False,countplot=False,boxplot=True,factorplot=False):
    data=data.copy()
    plt.figure(figsize=(14,9))
    if type(data[x].values[1])==str:
        from sklearn.preprocessing import LabelEncoder
        le=LabelEncoder()
        data[x]=data[x].apply(lambda y:treat_cat(y))
        data[x]=le.fit_transform(data[x])
    else:
        me=np.mean(data[x])
        data[x]=data[x].apply(lambda y: treat_num(y,m=me))
    try:
        print(name,'\n')
        data[x].hist()
        plt.show()
    except:
        print('cannot plot histogram\n')
    
    try:
        print(name,'\n')
        sns.distplot(train[x])
        plt.show()
    except:
        print('cannot plot distplot\n')
        
    if barplot:
        try:
            print(name,'\n')
            sns.barplot(x=x,y=y,hue=hue,data=data)
            plt.xlabel(x)
            plt.ylabel(y)
            plt.show()
        except:
            print('cannot plot barplot\n')
    if countplot:        
        try:
            print(name,'\n')
            sns.countplot(data[x])
            plt.xlabel(x)
            plt.ylabel(y)
            plt.show()
        except:
            print('cannot plot countplot\n')
    if boxplot:        
        try:
            print(name,'\n')
            sns.boxplot(x=x,y=y,hue=hue,data=data)
            plt.xlabel(x)
            plt.ylabel(y)
            plt.show()
        except:
            print('cannot plot boxplot\n')  
    if factorplot:        
        try:
            print(name,'\n')
            data=pd.DataFrame(data)
            sns.factorplot(x=x,y=y,data=data,size=14)
            plt.xlabel(x)
            plt.ylabel(y)
            plt.show()
        except:
            print('cannot plot factorplot\n')  
            
def plotHeatMap(df, size=(16,9),an=True):
    plt.figure(figsize=size)
    cmap = plt.cm.viridis
    sns.heatmap(df.corr(),cmap=cmap,linewidths=0.2,vmax=1, linecolor="white", square=True,annot=an)
    plt.show()
    
def treat_cat(tmp):
    if tmp!=tmp:
        return 'unknown'
    else:
        return tmp
    
def max_min(data):
    data=pd.DataFrame(data)
    print('max = '+str(max(data.values.ravel()))+'  min = '+str(min(data.values.ravel())))

def data_type(data):
    data=pd.DataFrame(data)
    for i in data.values.ravel():
        if i==i:
            return type(i)

			
# ======================================================================================================
# ======================================================================================================
#**************************************************************************************************************************	

#	best method for univariate analysis			

def describe(data,x=None, y=None, with_plot=True):
    data = data.copy()
    null_values(data[x])
    #for uniqueness
    try:
        if len(set(data[x].values))==len(data[x].values):
            print('is_unique : Yes')
        else:
            print('is_unique : No')
    except:
        pass
    
    no_of_unique_values(data[x])
    if len(set(data[x].values.ravel()))<=10:
        print(set(data[x].values.ravel()))
        count(data=data[x])
        
    if data_type(data[x])==str:
        from sklearn.preprocessing import LabelEncoder
        le=LabelEncoder()
        data[x]=data[x].apply(lambda y:treat_cat(y))
        data[x]=le.fit_transform(data[x])
    else:
        me=np.mean(data[x])
        data[x]=data[x].apply(lambda y: treat_num(y,m=me))
    data[x]=pd.DataFrame(data[x])
    max_min(data[x])
    non_zero(data[x])
    print(data[x].describe(),'\n')
    if with_plot:
        plot(x=x,data=data,y=y)



# ======================================================================================================
# ======================================================================================================
#**************************************************************************************************************************	
		
def non_zero(data):
    data=pd.DataFrame(data)
    print('non zero values = '+str(len(data[data.values.ravel()!=0]))+
         '   zero values = '+str(len(data[data.values.ravel()==0])))

def drop(data=None,col=None):
    return data.drop([i for i in col],axis=1)
    
def null_values(data):
     print('null values = '+str(len(data[data!=data])))
        
def treat_num(x,m):
    if x!=x:
        return m
    else:
        return x

def no_of_unique_values(data):
    print('Number of unique values : ',len(set(data)))
    
def count(data):
    from collections import Counter
    print(Counter(data))
    
def consider_unknown(data,li):
    if data in li:
        return data
    else :
        return 'unknown'

```

<h2>Preprocessing</h2>

```python
# So that was a little bit expolration
# but thinking practically,we can conclude the fact
# that all channels are dependent on each other
# as if a doctor is very much satisfied by one channel
# then he should most probably have low affinity towards 
# other channel.
# so we will create some features that will provide us
# total, max and avg affinity of a doctor from all 8 channels

# One Hot Encoding
from sklearn.preprocessing import OneHotEncoder
# Label encoding
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


# In histplot for Age, distribution seems to have a
# good volume of all aged doctors
# So, lets create a feature on the basis of
# carier experience, It seems setting <50 as low
#  >50 and <70 as Medium
# >70 as High experience would be good as per distribution

def get_experience(x):
    if x<50:
        return 'low'
    elif x>50 and x<70:
        return 'medium'
    else:
        return 'high'



def get_train_test_valid_data(data=None,col = None, test_size=0.2, ignore_cols_to_drop = []):
    data = data.copy()
        
    # simply filling gender with modes as only 28 missing values
    data['gender'] = data['gender'].fillna('MALE')
    
    # fill all nan's with mean other than given col to show its 
    # mean contribution, Note that as we are only using aggregating features,
    # so, mean wouldn't be having that much impact of biasing and as we
    # proceed to ignore some cols during mean imputation because at each step, we will be having true values
    # for one new missing col each time, biasing wouldn't be a problem :)
    for i in [j for j in ['DMS','DRT','RL','RR','DEM','OLA','OLV','P2P'] if j != col and j not in ignore_cols_to_drop]:
        data[i] = data[i].fillna(np.mean(all_data[all_data[i]==all_data[i]][i]))
    
    # feature generation
    # lets create some features using aggregation but column wise :) as all channels are dependent on each other
    # init
    data['total_channel_impact'] = 0
    
    for i in ['DMS','DRT','RL','RR','DEM','OLA','OLV','P2P']:
        if i!=col:
            data['total_channel_impact'] += data[i]
    data['avg_channel_impact'] = data['total_channel_impact']/7
    data['max_channel_impact'] = data[[i for i in ['DMS','DRT','RL','RR','DEM','OLA','OLV','P2P'] if i != col]].max(axis=1)
    
    # getting experience feature
    data['experience'] = data['Age'].apply(lambda x : get_experience(x))
    
    # Now let's normalize the Age feature with std scaler
    # to have 0 mean and unit variance
    data['Age'] = StandardScaler().fit_transform(data['Age'].reshape(-1,1)) 
    
    
    # getting rid of unneccessary dependent features
    data.drop([j for j in ['DMS','DRT','RL','RR','DEM','OLA','OLV','P2P'] if j != col],1,inplace=True)
    
    
    # Now lets one hot encode the categorical values
    cols_for_ohe = ['Region','Value','Speciality_ID','gender','experience']
    
    #Preprocess before OHE
    for i in cols_for_ohe:
        le = LabelEncoder()
        le.fit(data[i])
        data[i] = le.transform(data[i])
        
    
    # print('Started one hot encoding')
    for i in cols_for_ohe:
        # print(i,'\n',set(data[i]))

        ohe=OneHotEncoder(sparse =False)
        ohe.fit(data[i].values.reshape(-1,1))
        tmp=ohe.transform(data[i].values.reshape(-1,1))
        tmp=pd.DataFrame(tmp,dtype=int)
        tmp.columns=[i+'_'+str(j) for j in range(tmp.shape[1])]
        data=data.drop([i],1)
        data=pd.concat([data,tmp],1)        
        # print(data.shape)
    # print('Completed One Hot Encoding')
    
    # getting test data ie all those which are nan's in given col
    test_data = data.loc[(data[col]!=data[col]),[i for i in data.columns]]
    
    # get remaining non nan's data in a given col 
    tmp_data = data.loc[(data[col]==data[col]),[i for i in data.columns]]
    
    # note that, below i am creating train and validation data with given test_size
    # this
    train_data,validation_data,y_tr,y_ts = train_test_split(tmp_data,tmp_data[col],test_size=test_size,random_state=41526)
    
    return train_data,test_data,validation_data
```

```python
# DMS imputation

train,test,validation = get_train_test_valid_data(data=all_data,col='DMS',ignore_cols_to_drop=[])

dtrain = xgb.DMatrix(data=train.drop(['HCP_ID','DMS'],1), label=train['DMS'])
dvalid = xgb.DMatrix(data=validation.drop(['HCP_ID','DMS'],1), label=validation['DMS'])
dtest = xgb.DMatrix(data=test.drop(['HCP_ID','DMS'],1))

params = {}
params['objective'] = 'reg:linear'
params['eta'] = 0.01
params['silent'] = True
params['max_depth'] = 6
params['subsample'] = 0.8
params['colsample_bytree'] = 0.9
params['nthread']=-1
params['eval_metric']='rmse'

watchlist = [(dtrain, 'train'),(dvalid, 'eval')]

model1_1 = xgb.train(params, dtrain, 20000, watchlist, maximize=False, verbose_eval=50, early_stopping_rounds=260)

```

```python
#	Plotting feature importance through random forest

def plot_importance(model,train, top_n = None):
	# Classifier model must be pretrained
    plt.figure(figsize=(12,20))
    importance=model.feature_importances_
    features=np.array([str(i) for i in train.columns])
    indices = np.argsort(importance)
    indices = np.array([int(i)  for i in indices[(len(indices)-top_n):]])
    plt.title('Feature Importances')
    plt.barh(range(top_n), importance[indices], color='r', align='center')
    plt.yticks(range(len(indices)), features[indices])
    plt.xlabel('Relative Importance')
    plt.show()
```

```python
# Defining a imputer function

def get_missing_values(model = None, data = None, col = None, ignore_cols_to_drop = []):
    data = data.copy()
    
    train,test,validation = get_train_test_valid_data(data=all_data,col=col,ignore_cols_to_drop=ignore_cols_to_drop)
    model.fit(train.drop(['HCP_ID',col],1),train[col])
    print(col,'  - > ',np.power(mean_squared_error(y_true=validation[col],y_pred=model.predict(validation.drop(['HCP_ID',col],1))),.5))
    
    test[col] = model.predict(test.drop(['HCP_ID',col],1))
    return test[['HCP_ID',col]]

```

```python
sub = pd.read_csv('submission.csv')
cols_to_ignore = []

#  Filling missing values in increasing order of no of missing items, columns wise
for col in ['DMS','DRT','RL','RR','DEM','OLA','OLV','P2P']:
    # initializing model
    # Choosing parameters such that there will be many estimators overfitting in 
    # different reason which will most likely capture most of the patterns and then
    # bagging all of them with RandomForest
    model = RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=None,
               max_leaf_nodes=None,
               min_impurity_decrease=0.0, min_impurity_split=None,
               min_samples_leaf=50, min_samples_split=2,
               min_weight_fraction_leaf=0.0, n_estimators=1000, n_jobs=-1,
               oob_score=False, random_state=573869792, verbose=0, warm_start=False)

    result = get_missing_values(model=model,data=all_data,col=col,ignore_cols_to_drop=cols_to_ignore)
    sub.loc[sub['HCP_ID'].isin(result['HCP_ID'].values),col] = result[col].values
    
    # updating the data with new imputation
    cols_to_ignore += [col]
    all_data.loc[all_data['HCP_ID'].isin(result['HCP_ID'].values),col] = result[col].values
    
    
```
